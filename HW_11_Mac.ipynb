```python
import numpy as np
from astropy import constants as const
import matplotlib.pyplot as plt
```

This notebook is goint to be used to show the use of $r \propto (1-ecos(\theta)^{-1}$  in order to better understand the law through graphs by changing its parameters

```python
def r(e, θ):
    return (1-e*np.cos(θ))**(-1)
```

```python
θ= np.arange(0,2*np.pi, .01)
```

```python
r0 = r(0, θ)
r5 = r(.5,θ)
r9 = r(.9,θ)
r1 = r(1,θ)
r2 = r(2,θ)
r10 = r(10,θ)
```

```python



r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(θ, r0)
ax.set_rmax(1.5)
ax.set_rticks([0.5, 1, 1.5])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot of an orbit with e=0", va='bottom')
plt.show()
```
