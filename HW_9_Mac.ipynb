```python
# import numpy as np
import matplotlib.pyplot as plt
```

```python
def one(DE, k, T):
    return np.exp((-DE)/(k*T))
```

```python
DE = 1.65e-11
k = 1.381e-16
T = np.arange(1., 100000.)
plt.plot(one(DE, k, T))
```

```python
def two(Ne, me, kJ, T, hbar, Ei, kerg):
    return (2/Ne)*((me*kJ*T)/(2*np.pi*(hbar**2)))*np.exp(-Ei/(kerg*T))
```

```python
Ne = 10e26
me = 9.109e-31
kJ = 1.381e-23
T = np.arange(1., 100000.)
hbar = 1.055e-34
Ei = 2.2e-11
kerg = 1.381e-16
plt.plot(two(Ne, me, kJ, T, hbar, Ei, kerg))
```

```python
def three(DE, kerg, T, Ne, me, kJ, hbar, Ei):
    return (np.exp((-DE)/(kerg*T)))*(1/(1+((2/Ne)*((me*kJ*T)/(2*np.pi*(hbar**2)))*np.exp(-Ei/(kerg*T)))))
```

```python
dE = 1.65e-11
kerg = 1.381e-16
T = np.arange(1., 100000.)
Ne = 10e26
me = 9.109e-31
kJ = 1.381e-23
hbar = 1.055e-34
Ei = 2.2e-11
plt.plot(three(dE, kerg, T, Ne, me, kJ, hbar, Ei))
```

```python
plt.plot(one(DE, k, T))
plt.plot(two(Ne, me, kJ, T, hbar, Ei, kerg))
plt.plot(three(dE, kerg, T, Ne, me, kJ, hbar, Ei))
plt.show
```
